```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("homeprices.csv")
df.head()
```

```python
dummies = pd.get_dummies(pd.Series(list(df.town)), dtype=int)
dummies
```

```python
merged = pd.concat([df, dummies], axis=1)
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis=1)
final
```

```python
from sklearn.linear_model import LinearRegression
model  = LinearRegression()
```

```python
X = final.drop('price', axis=1)
X
```

```python
y = final.price
y
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400, 0,0]])
```

```python
model.score(X,y)
```
